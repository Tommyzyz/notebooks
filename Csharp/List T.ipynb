# 数据结构的使用

```csharp
//List<T>
 List<int> list=new List<int>();
 Console.WriteLine($"{list.Count}/{list.Capacity}");
 Console.WriteLine(String.Join(",",list));
 //默认构造数组为空数组
```

```csharp
int[] array=new int[]{100,200,300,400};
Console.WriteLine(array is IEnumerable<int>);
//有参构造，使用数组
list=new List<int>(array);
Console.WriteLine($"{list.Count}/{list.Capacity}");
Console.WriteLine(String.Join(",",list));
```

```csharp
//有参构造，使用集合
List<int> list2=new List<int>(list);
Console.WriteLine($"{list2.Count}/{list2.Capacity}");
Console.WriteLine(String.Join(",",list2));
```

```csharp
//有参构造，初始数组长度
List<int> list3=new List<int>(20);
Console.WriteLine($"{list3.Count}/{list3.Capacity}");
Console.WriteLine(String.Join(",",list3));
```

```csharp
//Add方法
//初始化长度4,8,16,32
 List<int> list4=new List<int>();
 for(int i=1;i<=100;i++)
 {
    list4.Add(i);
    Console.WriteLine($"{list4.Count}/{list4.Capacity}");
    Console.WriteLine(String.Join(",",list4));
 }
```

```csharp
//拥有Add方法即表明有初始化器
 List<int> list5=new List<int>(){100,200,300};
 list5.Add(20);
 Console.WriteLine($"{list5.Count}/{list5.Capacity}");
 Console.WriteLine(String.Join(",",list5));
 
```

```csharp
//AddRange方法 追加集合
 List<int> list6=new List<int>{10,20,30};
 list6.AddRange(list5);
 Console.WriteLine($"{list6.Count}/{list6.Capacity}");
 Console.WriteLine(String.Join(",",list6));
```

```csharp
 //Insrt方法 插入
 List<int> list7=new List<int>{10,20,30};
 list7.Insert(2,1000);
 Console.WriteLine($"{list7.Count}/{list7.Capacity}");
 Console.WriteLine(String.Join(",",list7));
 //模拟追加
 list7.Insert(list7.Count,1001);
 Console.WriteLine($"{list7.Count}/{list7.Capacity}");
 Console.WriteLine(String.Join(",",list7));
```

```csharp
 //InsrtRange方法 插入集合 位置，集合
 List<int> list8=new List<int>{0,1,2,3};
 list8.InsertRange(2,list7);
 Console.WriteLine($"{list8.Count}/{list8.Capacity}");
 Console.WriteLine(String.Join(",",list8));
```

```csharp
//Clear 只清数据,不缩长度
 list8.Clear();
 Console.WriteLine($"{list8.Count}/{list8.Capacity}");
```

```csharp
//RemoveAt
 List<int> list9=new List<int>{100,200,300,400,500,600};
 Console.WriteLine(String.Join(",",list9));
 list9.RemoveAt(0);
 Console.WriteLine(String.Join(",",list9));
 
 //RemoveRange 位置，元素数
 list9.RemoveRange(1,2);
 Console.WriteLine(String.Join(",",list9));
```

```csharp
//Remove方法 去掉第一个匹配值
 List<int> list10=new List<int>{100,400,300,400,500,300};
 Console.WriteLine(String.Join(",",list10));
 list10.Remove(400);
 Console.WriteLine(String.Join(",",list10));

//RemoveAll方法 去掉所有匹配值 参数为委托你好 
//去掉300
list10.RemoveAll(e=>e.Equals(300));
Console.WriteLine(String.Join(",",list10));
//去掉
list10.RemoveAll(e=>e%200==0);
Console.WriteLine(String.Join(",",list10));
```
