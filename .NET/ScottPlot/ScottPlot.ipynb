```csharp
// Install the ScottPlot NuGet package
#r "nuget:ScottPlot"

// Setup a custom formatter to display plots as images
using Microsoft.DotNet.Interactive.Formatting;
Formatter.Register(typeof(ScottPlot.Plot), (p, w) => 
    w.Write(((ScottPlot.Plot)p).GetSvgHtml(400, 300)), HtmlFormatter.MimeType);
```

```csharp
// create sample data
double[] dataX = [ 1, 2, 3, 4, 5 , 7];
double[] dataY = [ 1, 4, 9, 16, 25, 49];

// plot the data
ScottPlot.Plot plt = new();
plt.Add.Scatter(dataX, dataY);

// display the plot
plt
```

```csharp
#r "nuget:ScottPlot"
open ScottPlot;

let dataX = [| 1.0 .. 5.0 |]
let dataY = [| for x in dataX -> x * x |]
let plt = new Plot();
plt.Add.Scatter(dataX, dataY);
plt.SavePng("./图/quickstart.png", 400, 300);
plt
```
